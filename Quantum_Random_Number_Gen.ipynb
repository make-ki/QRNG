Quantum Source & Entropy Extraction:

1 . Hadamard Gate + Measurement	Creates superposition (50% |0⟩, 50% |1⟩); measurement collapses state for fundamental randomness ideal for simulation.

2. Beam Splitter with Single Photons	Photon's path (reflection/transmission) is inherently random; models real optical QRNGs
3. Vacuum Fluctuation Measurement	Measures inherent quantum noise in a vacuum; very high entropy and fast generation rates

There is a fundamental proof of complexity theory which states that it cannot be proved that a particular sequence of numbers is genuinely unpredictable.

```python
import qiskit
from qiskit.circuit.library import HGate
from qiskit_aer import AerSimulator

# I'll try to take noise from hadamard gates.
circuit = qiskit.QuantumCircuit(1)
circuit.append(HGate(), [0])
circuit.measure_all()

# No. of bits I want
N = int(input(f'Number of random bits needed: '))
random_num = []

# just simulating it on Aer and printing results
sampler = AerSimulator()
print(f'Starting {N} iterations...')
for i in range(0, N):
  job = sampler.run(circuits=[circuit], shots=1)
  result = job.result()
  dict_count = result.results[0].data.counts
  if dict_count.get('0x0',0) > dict_count.get('0x1', 0):
    random_num.append('0')
  else:
    random_num.append('1')
  print(f'{i} {result.results[0].data.counts}')
print(f'Iterations ended....')
bits = ''.join(random_num)
print(f'Random number generated : {bits}')
circuit.draw(output='mpl')
```

```python
def analyze_random_sequence(bits):
    n = len(bits)
    bits_list = [int(bit) for bit in bits]

    # Bias
    p = sum(bits_list) / n
    bias = p - 0.5

    # Convert to ±1
    x = [2*b - 1 for b in bits_list]

    # Autocorrelation at lag 1
    # We use the formula: r_1 = (1/(n-1)) * sum_{i=0}^{n-2} x_i * x_{i+1}
    if n > 1:
        r1 = sum(x[i] * x[i+1] for i in range(n-1)) / (n-1)
    else:
        r1 = 0

    return bias, r1

analyze_random_sequence(bits)
```

Von Neumann created a function that extracted entropy.
The drawback of his function is that if the distibution is already somewhat uniform we'll now have a random string that is much smaller than input and still somewhat uniform.

```python
def von_neumann_extractor(bit_string, output_type='string'):
    bits = [int(bit) for bit in bit_string]
    extracted_bits = []
    i = 0
    discarded_pairs = 0
    total_pairs = 0

    while i < len(bits) - 1:
        total_pairs += 1
        bit1, bit2 = bits[i], bits[i+1]

        # Von Neumann rule: 01 → 0, 10 → 1, 00/11 → discard
        if bit1 == 0 and bit2 == 1:
            extracted_bits.append(0)
        elif bit1 == 1 and bit2 == 0:
            extracted_bits.append(1)
        else:
            discarded_pairs += 1
            # No output for 00 or 11

        i += 2

    # Calculate efficiency metrics
    input_length = len(bits)
    output_length = len(extracted_bits)
    efficiency = output_length / input_length if input_length > 0 else 0

    print(f"Von Neumann Extractor Results:")
    print(f"Input bits: {input_length}")
    print(f"Output bits: {output_length}")
    print(f"Efficiency: {efficiency:.3f} ({output_length}/{input_length})")
    print(f"Discarded pairs: {discarded_pairs}/{total_pairs}")

    # Return in requested format
    if output_type == 'string':
        return ''.join(str(bit) for bit in extracted_bits)
    else:
        return extracted_bits

def analyze_extraction_effectiveness(original_bits, extracted_bits):
    def calculate_bias(bit_list):
        if len(bit_list) == 0:
            return 0, 0
        ones = sum(bit_list)
        proportion = ones / len(bit_list)
        bias = abs(proportion - 0.5)
        return bias, proportion

    orig_bias, orig_prop = calculate_bias(original_bits)
    extr_bias, extr_prop = calculate_bias(extracted_bits)

    print(f"\nBias Analysis:")
    print(f"Original - Proportion of 1s: {orig_prop:.4f}, Bias: {orig_bias:.4f}")
    if len(extracted_bits) > 0:
        print(f"Extracted - Proportion of 1s: {extr_prop:.4f}, Bias: {extr_bias:.4f}")
        bias_reduction = orig_bias - extr_bias
        print(f"Bias reduction: {bias_reduction:.4f}")
    else:
        print("No bits extracted!")


print(von_neumann_extractor(bits))
```

Its not reliable, variable length and information loss is a big downside, thats why you wouldn't see it in use anywhere. It's just shown cuz its simple. Now what is a function that can result fixed length output, changes significantly even for small changes in input?

```python
import numpy as np

class ToeplitzExtractor:
    """
    A simple Toeplitz hash-based randomness extractor.
    """

    def __init__(self, output_length, seed=None):
        """
        Initializes the extractor.

        Args:
            output_length (int): The desired number of bits in the final output.
            seed (int, optional): A seed for the random number generator to ensure
                                  reproducible keys. If None, the key is random.
        """
        self.output_length = output_length
        # The input length will be set when `extract` is called.
        self.input_length = None

        # Initialize a random number generator for creating the Toeplitz key
        self.rng = np.random.default_rng(seed)

    def _generate_toeplitz_matrix(self, n, m):
        """
        Generates a binary Toeplitz matrix of size (m x n).
        The first column and the first row define the entire matrix.

        Args:
            n (int): Number of columns (input length).
            m (int): Number of rows (output length).

        Returns:
            np.array: A binary (m x n) matrix representing the Toeplitz matrix.
        """
        # Total number of unique bits needed for an (m x n) Toeplitz matrix is (m + n - 1)
        total_bits = m + n - 1
        key = self.rng.integers(0, 2, size=total_bits)

        # Construct the matrix
        matrix = np.zeros((m, n), dtype=int)
        for i in range(m):
            for j in range(n):
                matrix[i, j] = key[n - 1 + i - j]
        return matrix

    def extract(self, input_bits):
        """
        Extracts a uniform random bit string from the input_bits using Toeplitz hashing.

        Args:
            input_bits (list or np.array): A 1D array-like object containing bits (0s and 1s).

        Returns:
            list: A list of bits (0s and 1s) of length `self.output_length`.
        """
        self.input_length = len(input_bits)
        input_vec = np.array(input_bits).reshape(-1, 1)  # Convert to column vector

        # Generate the Toeplitz matrix (acting as the universal hash function)
        T = self._generate_toeplitz_matrix(self.input_length, self.output_length)

        # Perform matrix multiplication in GF(2) (equivalent to XOR and AND)
        output_vec = (T @ input_vec) % 2
        return output_vec.flatten().tolist()

# Prepare the raw bits from the Hadamard gate output
raw_bits = [int(b) for b in bits]

# Initialize the extractor to get N bits out
extractor = ToeplitzExtractor(output_length=N, seed=42)

# Extract the final random bits
final_bits = extractor.extract(raw_bits)
final_bits_str = ''.join([chr(x+ord('0')) for x in final_bits])

print(f"Raw bits: {bits} of length {len(bits)}")
print(f"Final bits: {final_bits_str} of length {len(final_bits_str)}")
```

## NIST SP 800-22 Frequency (Monobit) Test

The NIST Statistical Test Suite (SP 800-22) is a collection of tests used to evaluate the randomness of binary sequences generated by cryptographic or quantum random number generators (QRNGs). The Frequency (Monobit) Test is one of the simplest and most fundamental tests.

### Purpose:

To check whether the number of 0s and 1s in a sequence are approximately equal. A large imbalance suggests the sequence is biased, meaning it may not be truly random.

### In this project:

We apply this test to the quantum-generated bits obtained using the Hadamard gate in Qiskit. It helps verify that the QRNG produces unbiased, random-looking data.

### Key Points:

*   **Test statistic:** compares counts of 0s and 1s.
*   Uses the **p-value** to decide if the sequence is random (p ≥ 0.01 → passes).
*   It’s the first step in validating QRNG output before applying more complex NIST tests.

```python
import math

def frequency_test(bits):
    n = len(bits)
    # Convert bits to integers and count number of 1s
    count = sum(int(b) for b in bits)
    # Compute test statistic (S_obs)
    s_obs = abs(count - (n - count)) / math.sqrt(n)
    # Compute p-value using complementary error function
    p_value = math.erfc(s_obs / math.sqrt(2))

    # Interpret result
    if p_value >= 0.01:
        result = "Sequence is random (passes test)"
    else:
        result = "Sequence is non-random (fails test)"

    print(f"Frequency Test Result:")
    print(f"  n = {n}")
    print(f"  Number of 1s = {count}")
    print(f"  S_obs = {s_obs:.4f}")
    print(f"  p-value = {p_value:.6f}")
    print(f"  Conclusion: {result}")

    return p_value, result

# Run the test on your quantum bits
p_value, result = frequency_test(bits)
```

 Runs Test

Purpose: Checks whether the bits alternate between 0 and 1 too frequently or too rarely.
Key Points:

Measures how often runs (consecutive identical bits) occur.

A balanced random sequence should have an expected number of runs.

Too many or too few runs → sequence may not be random

Screenshot 2025-11-07 151319.png

```python
import math

def runs_test(bits):
    n = len(bits)
    # Add a check for minimum sequence length
    if n < 100:
        print(f"Runs Test:\n  Sequence is too short (n = {n}). Minimum length required is 100 for this test.")
        return None, "Sequence too short for valid test"

    pi = sum(int(b) for b in bits) / n  # proportion of 1s

    # Check if sequence is too biased
    if abs(pi - 0.5) >= (2 / math.sqrt(n)):
        print("Runs Test:\n  Sequence is too biased (fails precondition).")
        return 0, "Sequence is non-random (fails test)"

    # Count total number of runs (changes between 0 and 1)
    runs = 1 + sum(bits[i] != bits[i - 1] for i in range(1, n))

    # Compute test statistic
    expected_runs = 2 * n * pi * (1 - pi)
    z = abs(runs - expected_runs) / (2 * math.sqrt(2 * n) * pi * (1 - pi))

    # Compute p-value
    p_value = math.erfc(z / math.sqrt(2))

    result = "Sequence is random (passes test)" if p_value >= 0.01 else "Sequence is non-random (fails test)"
    print(f"Runs Test:\n  n = {n}\n  π (proportion of 1s) = {pi:.4f}\n  Runs = {runs}\n  p-value = {p_value:.6f}\n  Conclusion: {result}")
    return p_value, result
```

Cumulative Sums (Cusum) Test

Purpose: Detects bias in the cumulative sum of bits over the sequence.
Key Points:

Treats each 1 as +1 and each 0 as -1, then computes cumulative sums.

Measures how far the sum strays from zero.

Large deviations suggest non-randomness or drift toward 0s or 1s.

Screenshot 2025-11-07 151424.png

```python
import math

def cusum_test(bits):
    n = len(bits)
    x = [1 if b == '1' else -1 for b in bits]
    s = [sum(x[:i+1]) for i in range(n)]
    z = max(abs(val) for val in s)
    p_value = 1 - sum(
        math.erfc((4*k*z - n) / (math.sqrt(2*n))) -
        math.erfc((4*k*z + n) / (math.sqrt(2*n)))
        for k in range((-n//(4*z)) + 1, (n//(4*z)) + 1)
    )

    result = "Sequence is random (passes test)" if p_value >= 0.01 else "Sequence is non-random (fails test)"
    print(f"Cumulative Sums Test:\n  Max deviation = {z}\n  p-value = {p_value:.6f}\n  Conclusion: {result}")
    return p_value, result
```
